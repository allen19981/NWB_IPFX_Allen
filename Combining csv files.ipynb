```python
import pandas as pd
df = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb.csv', index_col=0)
df1 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb1.csv', index_col=0)
df2 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb2.csv', index_col=0)
df3 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb3.csv', index_col=0)
df4 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb4.csv', index_col=0)
df5 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb5.csv', index_col=0)
df6 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb6.csv', index_col=0)
df7 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb7.csv', index_col=0)
df8 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb8.csv', index_col=0)
df9 = pd.read_csv ('SCC_runs/Mouse/mouse_features_from_nwb9.csv', index_col=0)

```
