```python
import pandas as pd
df = pd.read_csv ('SCC_runs/Human/human_features_from_nwb_local.csv', index_col=0)
df1 = pd.read_csv ('SCC_runs/Human/human_features_from_nwb_local1.csv', index_col=0)
df2 = pd.read_csv ('SCC_runs/Human/human_features_from_nwb_local2.csv', index_col=0)
df3 = pd.read_csv ('SCC_runs/Human/human_features_from_nwb_local3.csv', index_col=0)
df4 = pd.read_csv ('SCC_runs/Human/human_features_from_nwb_local4.csv', index_col=0)
```

```python
result = pd.concat([df, df1, df2, df3, df4], axis=1, sort=False)
```

```python
result.to_csv('full_human_features_from_nwb_local.csv')
```
